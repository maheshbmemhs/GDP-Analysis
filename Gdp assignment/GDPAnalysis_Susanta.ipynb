```python
# read the data
import pandas as pd
df=pd.read_csv(r"./State-wise_GDPatcurrentPriceonyearlybasis.csv")
```

```python
# Remove the rows for the year 2016-17
df = df.set_index("Duration")
df = df.drop('2016-17', axis=0)
df.reset_index(inplace=True)
```

```python
# filtering data 2013-14, 2014-15 ,2015-16 and '(% Growth over previous year)
df1 = df[ df['Items  Description'] =='(% Growth over previous year)']
durations=['2013-14','2014-15','2015-16']
df1=df1.loc[df['Duration'].isin(durations), :]
```

```python
#plotting the data to get the average growth rates of the various states:
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 5))
round(df1.mean(),2).sort_values().plot(kind='bar')
```

```python
# filtering total GDP,'2015-16'
df2 = df[ df['Items  Description'] =='GSDP - CURRENT PRICES (` in Crore)']
durations=['2015-16']
df2=df2.loc[df['Duration'].isin(durations), :]
#dropping columns
df2 = df2.drop('Duration', axis=1)
df2 = df2.drop('Items  Description', axis=1)
df2 = df2.drop('All_India GDP', axis=1)
df2=df2.iloc[0]
```

```python
#Plot the total GDP of the states for the year 2015-16
plt.figure(figsize=(12, 10))
df2.sort_values().plot(kind='barh')
```
