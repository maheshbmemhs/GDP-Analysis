#### Task4-IB: Plot the percentage contribution of primary, secondary and tertiary sectors as a percentage of total GDP for all the states.

`````python
#getting required rows for calculating percentage contribution of  Primary,Secondary,Tertiary to GSDP
df_merged_filtered_pst=df_merged_filtered.loc[(df_merged_filtered['Item']=='Primary') | (df_merged_filtered['Item']=='Secondary') | (df_merged_filtered['Item']=='Tertiary') | (df_merged_filtered['Item']=='Gross State Domestic Product')]
df_merged_filtered_pst.head(20)
`````

`````python
#Calculating Contribution Primary,Secondary,Tertiary Sectors to GSDP of State and inserting it in to dataframe
gsdp=df_merged_filtered_pst.loc[df_merged_filtered_pst['Item']=='Gross State Domestic Product',:]
count=0
count1=0
def func(num):
    global count,count1
    if count==4:
        count1+=1
        count=0
    count+=1              
    return num/gsdp.iloc[count1,0]
   

df_merged_filtered_pst['%Contribution']=round(100*df_merged_filtered_pst['2014-15'].apply(func))
`````

`````python
df_merged_filtered_pst.head(20)
`````

`````python
df_merged_filtered_pst=df_merged_filtered_pst[df_merged_filtered_pst['Item']!='Gross State Domestic Product']
`````

`````python
#pivoting the dataframe so that we can plot stacked bar graph
pivot_df_merged_filtered_pst=df_merged_filtered_pst.pivot(index='State',columns='Item',values='%Contribution')
pivot_df_merged_filtered_pst
`````

- Plot the percentage contribution of primary, secondary and tertiary sectors as a percentage of total GDP for all the states.

`````python

colors = ["yellow", "blue","green"]
pt=pivot_df_merged_filtered_pst.plot.bar(stacked=True, color=colors, figsize=(20,7))
pt.set_ylabel('%Contribution')
`````

#### Task5-IB: Categorise the states into four categories based on GDP per capita (C1, C2, C3, C4 - C1 would have the highest per capita GDP, C4 the lowest). The quantile values are (0.20,0.5, 0.85, 1), i.e. the states lying between the 85th and the 100th percentile are in C1, those between 50th and 85th percentile are in C2 and so on.

`````python
df_merged_filtered.head(33)
`````

`````python
#filtering out only Per Capita GSDP (Rs.)
df_merged_filtered_category=df_merged_filtered[df_merged_filtered['Item']=='Per Capita GSDP (Rs.)']
df_merged_filtered_category
`````

`````python
#Taking percentile based on 'Per Capita GSDP (Rs.)'
df_merged_filtered_category['Percentile']=round(100*df_merged_filtered_category['2014-15'].rank(pct=True))
df_merged_filtered_category
`````

`````python
#Categorizinng states in to c1 c2 c3 c4
def func(num):
    if num<=100 and num>=85:
        return 'C1'
    if num<85 and num>=50:
        return 'C2'
    if num<50 and num>=20:
        return 'C3'
    if num<20:
        return 'C4'
    
df_merged_filtered_category['Category']=df_merged_filtered_category['Percentile'].apply(func)
df_merged_filtered_category
`````

`````python
#matching the states of both dataframe so that they can be joined latter
def func(val):
    if val=='Arunachal_Pradesh':
        return 'Arunachal Pradesh'
    if val=='Andhra_Pradesh':
        return 'Andhra Pradesh'
    if val=='Himachal_Pradesh':
        return 'Himachal Pradesh'
    if val=='Tamil_Nadu':
        return 'Tamil Nadu'
    if val=='Uttar_Pradesh':
        return 'Uttar Pradesh'
    return val
        
df_merged_filtered_subsectors['State']=df_merged_filtered_subsectors['State'].apply(func)
df_merged_filtered_subsectors.head()
`````
